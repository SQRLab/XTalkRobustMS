```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_base = "MS_Gate_Parity_250us_Mode"
file_mid = "__0-50_"
file_end = "_Fidelity.npy"
file_end_error = "_Error.npy"
fid015 = np.load(file_base + "0_15kHz" + file_mid + "Target" + file_end)
fid115 = np.load(file_base + "1_15kHz" + file_mid + "Target" + file_end)
fid215 = np.load(file_base + "2_15kHz" + file_mid + "Target" + file_end)
fid015error = np.load(file_base + "0_15kHz" + file_mid + "Target" + file_end_error)
fid115error = np.load(file_base + "1_15kHz" + file_mid + "Target" + file_end_error)
fid215error = np.load(file_base + "2_15kHz" + file_mid + "Target" + file_end_error)

fid0neg15 = np.load(file_base + "0_neg15kHz" + file_mid + "Target" + file_end)
fid1neg15 = np.load(file_base + "1_neg15kHz" + file_mid + "Target" + file_end)
fid2neg15 = np.load(file_base + "2_neg15kHz" + file_mid + "Target" + file_end)
fid0neg15error = np.load(file_base + "0_neg15kHz" + file_mid + "Target" + file_end_error)
fid1neg15error = np.load(file_base + "1_neg15kHz" + file_mid + "Target" + file_end_error)
fid2neg15error = np.load(file_base + "2_neg15kHz" + file_mid + "Target" + file_end_error)

fid015XTalk = np.load(file_base + "0_15kHz" + file_mid + "XTalk" + file_end)
fid115XTalk = np.load(file_base + "1_15kHz" + file_mid + "XTalk" + file_end)
fid215XTalk = np.load(file_base + "2_15kHz" + file_mid + "XTalk" + file_end)
fid015XTalkerror = np.load(file_base + "0_15kHz" + file_mid + "XTalk" + file_end_error)
fid115XTalkerror = np.load(file_base + "1_15kHz" + file_mid + "XTalk" + file_end_error)
fid215XTalkerror = np.load(file_base + "2_15kHz" + file_mid + "XTalk" + file_end_error)

fid0neg15XTalk = np.load(file_base + "0_neg15kHz" + file_mid + "XTalk" + file_end)
fid1neg15XTalk = np.load(file_base + "1_neg15kHz" + file_mid + "XTalk" + file_end)
fid2neg15XTalk = np.load(file_base + "2_neg15kHz" + file_mid + "XTalk" + file_end)
fid0neg15XTalkerror = np.load(file_base + "0_neg15kHz" + file_mid + "XTalk" + file_end_error)
fid1neg15XTalkerror = np.load(file_base + "1_neg15kHz" + file_mid + "XTalk" + file_end_error)
fid2neg15XTalkerror = np.load(file_base + "2_neg15kHz" + file_mid + "XTalk" + file_end_error)

theory = np.load("Theory_Results.npy")
```

```python
Xtalk = [0,.1,.2,.3,.4,.5]
plt.plot(Xtalk, fid015-fid0neg15, label="0 diff",color='r')
plt.plot(Xtalk, fid115-fid1neg15, label="1 diff",color='b')
plt.plot(Xtalk, fid215-fid2neg15, label="2 diff",color='g')

plt.title("Contrast Difference between 15kHz and -15kHz")
plt.xlabel("Cross Talk")
plt.ylabel("Contrast Difference")
plt.legend()
```

```python
plt.plot(Xtalk, fid015XTalk-fid0neg15XTalk, label="0 diff",color='red')
plt.plot(Xtalk, fid115XTalk-fid1neg15XTalk, label="1 diff",color='blue')
plt.plot(Xtalk, fid215XTalk-fid2neg15XTalk, label="2 diff",color='green')
plt.title("Contrast Difference between 15kHz and -15kHz")
plt.xlabel("Cross Talk")
plt.ylabel("Contrast Difference")
plt.legend()
```

```python
plt.errorbar(Xtalk, fid015, yerr=fid015error, label="0 15kHz", color='red')
plt.errorbar(Xtalk, fid115, yerr=fid115error, label="1 15kHz", color='blue')
plt.errorbar(Xtalk, fid215, yerr=fid215error, label="2 15kHz", color='green')
plt.errorbar(Xtalk, fid0neg15, yerr=fid0neg15error, label="0 -15kHz", color='red', linestyle='dashed')
plt.errorbar(Xtalk, fid1neg15, yerr=fid1neg15error, label="1 -15kHz", color='blue', linestyle='dashed')
plt.errorbar(Xtalk, fid2neg15, yerr=fid2neg15error, label="2 -15kHz", color='green', linestyle='dashed')
plt.title("Contrast change of target pair with increasing Xtalk")
plt.xlabel("Cross Talk")
plt.ylabel("Contrast")
plt.legend()
```

```python
plt.errorbar(Xtalk, fid015XTalk, yerr=fid015XTalkerror, label="0 15kHz", color='red')
plt.errorbar(Xtalk, fid115XTalk, yerr=fid115XTalkerror, label="1 15kHz", color='blue')
plt.errorbar(Xtalk, fid215XTalk, yerr=fid215XTalkerror, label="2 15kHz", color='green')
plt.errorbar(Xtalk, fid0neg15XTalk, yerr=fid0neg15XTalkerror, label="0 -15kHz", color='red', linestyle='dashed')
plt.errorbar(Xtalk, fid1neg15XTalk, yerr=fid1neg15XTalkerror, label="1 -15kHz", color='blue', linestyle='dashed')
plt.errorbar(Xtalk, fid2neg15XTalk, yerr=fid2neg15XTalkerror, label="2 -15kHz", color='green', linestyle='dashed')
plt.title("Contrast change of non-target pairs with increasing Xtalk")
plt.xlabel("Cross Talk")
plt.ylabel("Contrast")
plt.legend()
```
